## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Name
summary_df = complete_df.groupby(['school_name'])
name_of_schools = list(complete_df.school_name.sort_values().unique())

# School Type
school_types = school_data.sort_values(by="school_name").type

# Total Students
student_count = list(summary_df.student_name.count())

# Total School Budget
school_budget = list(summary_df.budget.count())

# Per Student Budget
budget = list(summary_df.budget.mean())
budget_student = [i/j for i,j in zip(budget,student_count)]

# Average Math Score
school_math_avg = list(summary_df.math_score.mean())

# Average Reading Score
school_read_avg = list(summary_df.reading_score.mean())


# Calculating the passing math percentages per school %
summary = complete_df[complete_df['math_score']>=70].groupby(['school_name'])
math_summary = [(i/j)*100 for i,j in zip(summary.math_score.count(),student_count)]

# Calculating the passing reading percentages per school %
summary = complete_df[complete_df['reading_score']>=70].groupby(['school_name'])
read_summary = [(i/j)*100 for i,j in zip(summary.reading_score.count(),student_count)]

# % Overall Passing (The percentage of students that passed math and reading per school)
school_overall = [(i/j)*100 for i,j in zip(math_summary,read_summary)]

#Create a dataframe to hold the above results
School_Summary_df = pd.DataFrame({"School Name": name_of_schools,
                            "School type": school_types,
                            "Total Budget": school_budget,
                            "Per Student Budget": budget_student,                                                                                             
                            "Average Math Score": school_math_avg,
                            "Average Reading Score": school_read_avg, 
                            "% Passing Math": math_summary,
                            "% Passing Reading": read_summary,
                            "% Overall Passing": school_overall})


School_Summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
School_Summary_df.head()
```

```python
#copying the School_Summary_df in order to reuse DF
Spending_scores = School_Summary_df.copy()

# Create bins in which to place values based upon School Spend by student
bins = [0, 585, 615, 645, 675]

# Create labels for these bins
labels_by_spend = ['$0-585', '$585-615', '$616-645', '$646-675']


Spending_scores["Spending Ranges per student"] = pd.cut(Spending_scores["Per Student Budget"], bins, labels=labels_by_spend)

# Create a GroupBy object based upon "Spending Ranges per student"
Spending_scores = Spending_scores.groupby("Spending Ranges per student")

# Get the average of each column within the GroupBy object
Student_Ranges_df = pd.DataFrame(Spending_scores[["Spending Ranges per student", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean())

Student_Ranges_df
```
